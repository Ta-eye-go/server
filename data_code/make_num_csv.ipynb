```python
import csv

file = open('drive/MyDrive/removed_overlap_busnumber.csv', 'r', encoding='cp949')
rd = csv.reader(file)
next(rd)

li = []

for row in rd:
  li.append(row)

file.close()
print(li)
```

```python
print(li[1])
```

```python
print(li[1][0][:-1])
```

```python
#'번' 제거
for i in range(len(li)):
  if li[i][0][-1] == '번':
    li[i][0] = li[i][0][:-1]
print(li)
```

```python
# '번' 버스번호인지 확인
for i in range(len(li)):
  if li[i][0][-1] == '번':
    print(li[i])
```

```python
# 모두 제거
for i in range(len(li)):
  if li[i][0][-1] == '번':
    li[i][0] = li[i][0][:-1]
print(li)
```

```python
mod_file = open('drive/MyDrive/modified_removed_overlap_busnumber.csv', 'w', newline='', encoding='cp949')
wr = csv.writer(mod_file)
for i in range(len(li)):
  wr.writerow(li[i])

mod_file.close()
```

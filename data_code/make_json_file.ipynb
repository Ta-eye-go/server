```python
import openpyxl
from openpyxl import load_workbook
import json
from collections import OrderedDict
```

```python
rb = load_workbook("drive/MyDrive/filename.xlsx")     # file name 수정(엑셀파일)
ws = rb['Sheet1']
```

```python
r = 1
enti_list=[]
n = 65

while ws[f'A{r}'].value != None:
  syn_list=[]
  v = ws[f'A{r}'].value
  n = 66
  while (ws[f'{chr(n)}{r}'].value != None):
    syn_list.append(ws[f'{chr(n)}{r}'].value)
    n += 1
    
  enti_list.append({"value": v, "synonyms": syn_list})
  r += 1

print(enti_list)
```

```python
file_data = OrderedDict()
```

```python
file_data["name"] = "operation_system"
file_data["isOverridable"] = "true"
file_data["entries"] = enti_list
file_data["isEnum"] = "false"
file_data["automatedExpansion"] = "false"
```

```python
print(json.dumps(file_data, ensure_ascii=False, indent="\t"))
```

```python
with open('drive/MyDrive/words.json', 'w', encoding="utf-8") as make_file:      # file name 수정
    json.dump(file_data, make_file, ensure_ascii=False, indent="\t")
```
